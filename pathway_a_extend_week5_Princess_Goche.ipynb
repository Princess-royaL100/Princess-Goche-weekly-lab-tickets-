Step 5: Transform Your Core Functions

Convert your Week 5 functions to work with dictionaries:bold text

```python
# ============================================================================
# IMPORTS
# ============================================================================
from datetime import datetime

# ============================================================================
# TRACKER SETUP (DICTIONARY-BASED SYSTEM)
# ============================================================================

finance_tracker = {
    "user_info": {
        "name": "Demo User",
        "monthly_budget": 2500,
        "tracker_created": "2026-04-07"
    },
    "categories": {
        "Housing": {"budget": 1000, "spent": 0},
        "Food": {"budget": 500, "spent": 0},
        "Transport": {"budget": 300, "spent": 0},
        "Entertainment": {"budget": 300, "spent": 0},
        "Other": {"budget": 400, "spent": 0}
    },
    "statistics": {
        "total_expenses": 0,
        "expense_count": 0,
        "last_expense_id": 0
    },
    "expenses": {}
}

# ============================================================================
# ENHANCED EXPENSE FUNCTIONS (DICTIONARY VERSION)
# ============================================================================

def generate_expense_id(tracker):
    tracker["statistics"]["last_expense_id"] += 1
    return f"EXP{tracker['statistics']['last_expense_id']:03d}"


def add_expense_enhanced(tracker, amount, description, category="Other", essential=True):
    expense_id = generate_expense_id(tracker)

    expense = {
        "amount": amount,
        "description": description,
        "category": category,
        "date": datetime.now().strftime("%Y-%m-%d"),
        "essential": essential
    }

    tracker["expenses"][expense_id] = expense

    # Update category
    if category in tracker["categories"]:
        tracker["categories"][category]["spent"] += amount

    # Update stats
    tracker["statistics"]["total_expenses"] += amount
    tracker["statistics"]["expense_count"] += 1

    return expense_id


def get_expenses_by_category(tracker, category):
    return {
        eid: data for eid, data in tracker["expenses"].items()
        if data["category"] == category
    }


def calculate_category_summary(tracker):
    summary = {}

    for category, info in tracker["categories"].items():
        budget = info["budget"]
        spent = info["spent"]

        summary[category] = {
            "budget": budget,
            "spent": spent,
            "remaining": budget - spent,
            "percentage": (spent / budget * 100) if budget > 0 else 0,
            "expense_count": len(get_expenses_by_category(tracker, category))
        }

    return summary

# ============================================================================
# TRANSFORMED WEEK 5 FUNCTIONS (NOW USING DICTIONARY)
# ============================================================================

def calculate_total_spent(tracker):
    return tracker["statistics"]["total_expenses"]


def calculate_percentage(tracker):
    total = calculate_total_spent(tracker)
    budget = tracker["user_info"]["monthly_budget"]
    return (total / budget * 100) if budget > 0 else 0


def calculate_average_expense(tracker):
    count = tracker["statistics"]["expense_count"]
    total = tracker["statistics"]["total_expenses"]
    return (total / count) if count > 0 else 0


def find_largest_expense(tracker):
    if not tracker["expenses"]:
        return 0
    return max(exp["amount"] for exp in tracker["expenses"].values())


def find_smallest_expense(tracker):
    if not tracker["expenses"]:
        return 0
    return min(exp["amount"] for exp in tracker["expenses"].values())


def count_expenses_by_category(tracker):
    counts = {"Major": 0, "Minor": 0}

    for exp in tracker["expenses"].values():
        if exp["amount"] > 100:
            counts["Major"] += 1
        else:
            counts["Minor"] += 1

    return counts

# ============================================================================
# BONUS FUNCTIONS (UPDATED)
# ============================================================================

def calculate_savings_timeline(current, target, monthly_save):
    if monthly_save <= 0:
        return float('inf')

    remaining = target - current
    if remaining <= 0:
        return 0

    return remaining / monthly_save


def generate_detailed_advice(tracker):
    if not tracker["expenses"]:
        return "Start tracking expenses!"

    total = calculate_total_spent(tracker)
    percentage = calculate_percentage(tracker)
    average = calculate_average_expense(tracker)
    categories = count_expenses_by_category(tracker)

    advice = []

    if percentage >= 90:
        advice.append("🚨 Near budget limit!")
    elif percentage >= 75:
        advice.append("⚠️ Slow spending down.")
    elif percentage <= 25:
        advice.append("🌟 Great saving potential!")

    if categories["Major"] > 2:
        advice.append("💡 Too many large expenses.")

    if average > tracker["user_info"]["monthly_budget"] * 0.1:
        advice.append("💰 Average expense is high.")

    return " ".join(advice) if advice else "👍 Spending looks balanced!"

# ============================================================================
# TESTING SYSTEM
# ============================================================================

print("🧪 RUNNING UPDATED SYSTEM TEST\n")

# Add test expenses
add_expense_enhanced(finance_tracker, 850, "Rent payment", "Housing", True)
add_expense_enhanced(finance_tracker, 125, "Groceries", "Food", True)
add_expense_enhanced(finance_tracker, 65, "Gas", "Transport", True)
add_expense_enhanced(finance_tracker, 45, "Coffee", "Food", False)
add_expense_enhanced(finance_tracker, 200, "Utilities", "Housing", True)
add_expense_enhanced(finance_tracker, 75, "Movies", "Entertainment", False)

# Run calculations
total = calculate_total_spent(finance_tracker)
percentage = calculate_percentage(finance_tracker)
average = calculate_average_expense(finance_tracker)
largest = find_largest_expense(finance_tracker)
smallest = find_smallest_expense(finance_tracker)
categories = count_expenses_by_category(finance_tracker)

print("📊 RESULTS:")
print(f"Total: ${total:.2f}")
print(f"Budget %: {percentage:.1f}%")
print(f"Average: ${average:.2f}")
print(f"Largest: ${largest:.2f}")
print(f"Smallest: ${smallest:.2f}")
print(f"Categories: {categories}")

# Advice
print("\n💡 ADVICE:")
print(generate_detailed_advice(finance_tracker))

# Savings demo
months = calculate_savings_timeline(1500, 5000, 200)
print(f"\n🎯 Savings goal: {months:.1f} months")
```

### Step 6: Enhanced Analysis Functions

Create more sophisticated analysis using dictionary power:

```python
# ============================================================================
# IMPORT
# ============================================================================
from datetime import datetime

# ============================================================================
# ENHANCED ANALYSIS FUNCTIONS
# ============================================================================

def generate_financial_report(tracker):
    user_info = tracker["user_info"]
    stats = tracker.get("statistics", {"total_expenses": 0})
    monthly_budget = user_info["monthly_budget"]

    print("=" * 60)
    print(f"📊 FINANCIAL REPORT FOR {user_info['name'].upper()}")
    print("=" * 60)
    print(f"📅 Report Date: {datetime.now().strftime('%Y-%m-%d %H:%M')}")
    print(f"🗓️ Tracker Created: {user_info['tracker_created']}")
    print()

    # =======================
    # BUDGET OVERVIEW
    # =======================
    total_spent = stats["total_expenses"]
    budget_percentage = (total_spent / monthly_budget * 100) if monthly_budget > 0 else 0
    remaining_budget = monthly_budget - total_spent

    print("💰 BUDGET OVERVIEW:")
    print(f"   Monthly Budget: ${monthly_budget:.2f}")
    print(f"   Total Spent: ${total_spent:.2f}")
    print(f"   Remaining: ${remaining_budget:.2f}")
    print(f"   Budget Used: {budget_percentage:.1f}%")
    print()

    # =======================
    # CATEGORY BREAKDOWN
    # =======================
    print("📊 CATEGORY BREAKDOWN:")
    category_summary = calculate_category_summary(tracker)

    for category, info in category_summary.items():
        status_icon = "⚠️" if info["spent"] > info["budget"] else "✅"
        print(f"   {status_icon} {category}: ${info['spent']:.2f} / ${info['budget']:.2f} "
              f"({info['percentage']:.1f}%) - {info['expense_count']} expenses")
    print()

    # =======================
    # RECENT EXPENSES
    # =======================
    print("📝 RECENT EXPENSES:")
    recent_expenses = list(tracker.get("expenses", {}).items())[-5:]

    if recent_expenses:
        for expense_id, expense_data in recent_expenses:
            essential_icon = "🔴" if expense_data["essential"] else "🟡"
            print(f"   {expense_id}: ${expense_data['amount']:.2f} - {expense_data['description']} "
                  f"({expense_data['category']}) {essential_icon}")
    else:
        print("   No expenses recorded yet.")
    print()

    # =======================
    # SMART INSIGHTS
    # =======================
    print("💡 FINANCIAL INSIGHTS:")

    if budget_percentage >= 90:
        print("   🚨 You're close to your budget limit - reduce non-essential spending")
    elif budget_percentage >= 75:
        print("   ⚡ Approaching limit - monitor spending")
    elif budget_percentage >= 50:
        print("   📊 You're halfway - good control")
    else:
        print("   🌟 Excellent budget management")

    # Largest expense
    if tracker.get("expenses"):
        largest_expense = max(tracker["expenses"].items(), key=lambda x: x[1]["amount"])
        print(f"   💸 Largest expense: {largest_expense[0]} - ${largest_expense[1]['amount']:.2f} "
              f"({largest_expense[1]['description']})")

    print("=" * 60)


def find_optimization_opportunities(tracker):
    opportunities = []
    category_summary = calculate_category_summary(tracker)

    # =======================
    # OVESPENDING
    # =======================
    for category, info in category_summary.items():
        if info["spent"] > info["budget"]:
            overspend = info["spent"] - info["budget"]
            opportunities.append(
                f"Reduce {category} spending by ${overspend:.2f}"
            )

    # =======================
    # UNUSED BUDGET
    # =======================
    for category, info in category_summary.items():
        if info["expense_count"] == 0:
            opportunities.append(
                f"Reallocate unused ${info['budget']:.2f} from {category}"
            )

    # =======================
    # EXPENSIVE NON-ESSENTIAL ITEMS
    # =======================
    threshold = tracker["user_info"]["monthly_budget"] * 0.05

    for expense_id, expense_data in tracker.get("expenses", {}).items():
        if expense_data["amount"] >= threshold and not expense_data["essential"]:
            opportunities.append(
                f"Review {expense_id}: ${expense_data['amount']:.2f} ({expense_data['description']})"
            )

    return opportunities


# ============================================================================
# TESTING
# ============================================================================

print("🧪 Testing enhanced analysis:\n")

finance_tracker["user_info"]["name"] = "Demo User"

# Run report
generate_financial_report(finance_tracker)

# Optimization insights
opportunities = find_optimization_opportunities(finance_tracker)

print("\n🎯 OPTIMIZATION OPPORTUNITIES:")
if opportunities:
    for i, opp in enumerate(opportunities, 1):
        print(f"   {i}. {opp}")
else:
    print("   ✅ No optimization opportunities found - excellent financial management!")
```

## 🎯 Phase 3: Progressive Enhancement (30 minutes)

### Step 7: Add Advanced Dictionary Features

Implement sophisticated features that would be difficult with lists:

```python
# ============================================================================
# ADVANCED DICTIONARY FEATURES
# ============================================================================

# ============================================================================
# IMPORT
# ============================================================================
from datetime import datetime

# ============================================================================
# ADVANCED DICTIONARY FEATURES
# ============================================================================

def create_expense_filter(tracker, **criteria):
    filtered_expenses = {}

    for expense_id, expense_data in tracker.get("expenses", {}).items():
        include_expense = True

        for criterion, value in criteria.items():

            if criterion == "category" and expense_data.get("category") != value:
                include_expense = False
                break

            elif criterion == "essential" and expense_data.get("essential") != value:
                include_expense = False
                break

            elif criterion == "min_amount" and expense_data.get("amount", 0) < value:
                include_expense = False
                break

            elif criterion == "max_amount" and expense_data.get("amount", float('inf')) > value:
                include_expense = False
                break

            elif criterion == "date" and expense_data.get("date") != value:
                include_expense = False
                break

            elif criterion == "description_contains":
                if value.lower() not in expense_data.get("description", "").lower():
                    include_expense = False
                    break

        if include_expense:
            filtered_expenses[expense_id] = expense_data

    return filtered_expenses


def create_monthly_summary_by_date(tracker):
    monthly_summary = {}

    for expense_id, expense_data in tracker.get("expenses", {}).items():
        expense_date = expense_data.get("date", "")

        if expense_date:
            year_month = expense_date[:7]  # YYYY-MM

            if year_month not in monthly_summary:
                monthly_summary[year_month] = {
                    "total_amount": 0,
                    "expense_count": 0,
                    "categories": {},
                    "expenses": {}
                }

            monthly_summary[year_month]["total_amount"] += expense_data["amount"]
            monthly_summary[year_month]["expense_count"] += 1
            monthly_summary[year_month]["expenses"][expense_id] = expense_data

            category = expense_data.get("category", "Other")

            if category not in monthly_summary[year_month]["categories"]:
                monthly_summary[year_month]["categories"][category] = 0

            monthly_summary[year_month]["categories"][category] += expense_data["amount"]

    return monthly_summary


def export_expense_data(tracker, format="summary"):

    if format == "summary":
        return {
            "export_date": datetime.now().isoformat(),
            "user": tracker["user_info"]["name"],
            "total_budget": tracker["user_info"]["monthly_budget"],
            "total_spent": tracker["statistics"]["total_expenses"],
            "total_expenses": tracker["statistics"]["expense_count"],
            "category_summary": calculate_category_summary(tracker)
        }

    elif format == "detailed":
        return {
            "export_date": datetime.now().isoformat(),
            "complete_tracker": tracker
        }

    elif format == "csv_ready":
        csv_data = []

        for expense_id, expense_data in tracker.get("expenses", {}).items():
            csv_data.append({
                "expense_id": expense_id,
                "amount": expense_data["amount"],
                "description": expense_data["description"],
                "category": expense_data["category"],
                "date": expense_data["date"],
                "essential": expense_data["essential"]
            })

        return csv_data

    else:
        return {"error": "Invalid export format"}


# ============================================================================
# TESTING ADVANCED FEATURES
# ============================================================================

print("🧪 Testing advanced dictionary features:\n")

# Add more test data
add_expense_enhanced(finance_tracker, 25, "Coffee with friends", "Entertainment", False)
add_expense_enhanced(finance_tracker, 150, "Electric bill", "Utilities", True)
add_expense_enhanced(finance_tracker, 45, "Lunch at restaurant", "Food", False)

# =======================
# FILTERING TESTS
# =======================
print("🔍 FILTERING EXAMPLES:")

food_expenses = create_expense_filter(finance_tracker, category="Food")
print(f"Food expenses: {len(food_expenses)} found")

expensive_expenses = create_expense_filter(finance_tracker, min_amount=100)
print(f"Expensive expenses (>$100): {len(expensive_expenses)} found")

non_essential = create_expense_filter(finance_tracker, essential=False)
print(f"Non-essential expenses: {len(non_essential)} found")


# =======================
# MONTHLY SUMMARY
# =======================
monthly_data = create_monthly_summary_by_date(finance_tracker)

print("\n📅 MONTHLY SUMMARY:")
for month, data in monthly_data.items():
    print(f"   {month}: ${data['total_amount']:.2f} ({data['expense_count']} expenses)")


# =======================
# EXPORT TEST
# =======================
export_summary = export_expense_data(finance_tracker, "summary")

print("\n📤 EXPORT READY:")
print(f"   Total Budget: ${export_summary['total_budget']:.2f}")
print(f"   Total Spent: ${export_summary['total_spent']:.2f}")
print(f"   Export contains: {len(export_summary)} sections")
```
