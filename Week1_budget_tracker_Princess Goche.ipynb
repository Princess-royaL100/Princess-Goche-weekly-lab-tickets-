## Part 3: Building the Budget Tracker

Use AI to help you write a program that:
- Allows users to input their monthly income
- Allows users to input up to 3 expenses  
- Calculates remaining budget
- Shows a warning if budget is negative

### 💡 Pro Tip: Be specific about requirements!

```python
# TASK: Write your detailed prompt for the budget tracker
"""
My Budget Tracker Prompt:
write a python program that:

Allows users to input their monthly income
Allows users to input up to 3 expenses
Calculates remaining budget
Shows a warning if budget is negative
"""
```

### Paste and Test Your AI-Generated Budget Tracker

```python
# AI-GENERATED BUDGET TRACKER CODE
# Get monthly income from the user
income = float(input("Enter your monthly income: "))

# Create an empty list to store expenses
expenses = []

# Ask the user for up to 3 expenses
for i in range(1, 4):
    expense = float(input(f"Enter expense {i} (or 0 if none): "))
    expenses.append(expense)

# Calculate the total expenses
total_expenses = sum(expenses)

# Calculate the remaining budget
remaining_budget = income - total_expenses

# Display results
print(f"\nTotal income: {income}")
print(f"Total expenses: {total_expenses}")
print(f"Remaining budget: {remaining_budget}")

# Show a warning if the budget is negative
if remaining_budget < 0:
    print("⚠️ Warning: Your budget is negative! You are overspending.")
```

## Part 4: Enhancement Challenge

Choose ONE feature to add using AI assistance:
- 🏷️ Expense categories (food, transport, entertainment)
- 📊 Show percentage of income spent
- ✅ Input validation (no negative numbers)
- 📝 Detailed budget summary with advice

```python
# Document your enhancement
enhancement_choice = "I chose to add: a detailed budget Summary with advice"
enhancement_prompt = "I instructed the AI tool to add the detailed budget summary with advice to the previous code"
```

```python
# ENHANCED BUDGET TRACKER
# Get the user's monthly income
monthly_income = float(input("Enter your monthly income: "))

print("\nNow enter up to 3 monthly expenses.\n")
expenses = []

# Loop to input up to 3 expenses
for i in range(1, 4):
    expense_name = input(f"Enter name of expense #{i} (or press Enter to skip): ")
    if expense_name == "":
        break  # stop early if user doesn't want to enter more
    expense_amount = float(input(f"Enter amount for {expense_name}: "))
    expenses.append((expense_name, expense_amount))

# Calculate total expenses
total_expenses = sum(amount for _, amount in expenses)

# Calculate remaining budget
remaining_budget = monthly_income - total_expenses

# Print detailed summary
print("\n===== Budget Summary =====")
print(f"Monthly Income:        ${monthly_income:.2f}")
print(f"Total Expenses:        ${total_expenses:.2f}")

for name, amount in expenses:
    print(f"  - {name}:           ${amount:.2f}")

print(f"Remaining Budget:      ${remaining_budget:.2f}")

# Show warning if budget is negative
if remaining_budget < 0:
    print("\n⚠️  Warning: Your expenses are higher than your income!")
    print("Consider reducing some expenses or increasing your income.")
else:
    print("\n👍 You're within your budget.")
    if remaining_budget > (monthly_income * 0.2):
        print("Nice! You’re saving more than 20% of your income!")
```
